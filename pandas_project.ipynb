```python
# most common jon title
most_common_job_title="Platform Engineer"
print(f"most common job title is {most_common_job_title},{df_cleaned['job_title'].value_counts().max()}")
```

```python
most_common_job_title="Platform Engineer"
Platform_Engineer =df_cleaned[df_cleaned['job_title']==most_common_job_title]
Platform_Engineer
```

```python
PE_max_salary=Platform_Engineer['annual_salary_usd'].max()
print(PE_max_salary)
top_earner=Platform_Engineer[Platform_Engineer['annual_salary_usd']==PE_max_salary]
print(top_earner)
```

```python
oldest_person=df_cleaned['age'].max()
print(oldest_person)
olderst_employee=df_cleaned[df_cleaned['age']==oldest_person]
print(olderst_employee)
```

```python
# average salary by gender
avg_salary_by_gender=df_cleaned.groupby('gender')['annual_salary_usd'].max()
avg_salary_by_gender
```

```python
bachelor_degree_people=df_cleaned[df_cleaned['education_level']=="Bachelor's Degree"]
print(bachelor_degree_people)
```

```python
# find the age between 30-40
import numpy as np
age_30_40=df_cleaned[np.logical_and((df_cleaned['age']>=30), (df_cleaned['age']<=40))]
print(age_30_40)
age_30_40.shape
```

```python
# filter job title  aged between 30 to 40
abc=df_cleaned[df_cleaned['age'].between(30,40)][['job_title','age','first_name','location']]
abc
```

```python
# filter salary between 300000 to 400000
xyz=df_cleaned[df_cleaned['annual_salary_usd'].between(300000,400000)][['job_title','age','first_name','location']]
xyz
```
