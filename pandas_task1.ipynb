Write a Python program using Pandas to create a table that contains the following columns and data:
Employee Name: John, Kate, Leo
Department: IT, HR, Sales
Salary: 50000, 55000, 60000

for Both series and dataframe using dictionary and list of list

```python
import pandas as pd
employes={"employee_name":['John', 'Kate', 'Leo'], "department":['IT', 'HR', 'Sales'],"salary":[50000, 55000, 60000]}
employes_seres=pd.Series(employes)
print(employes_seres)
```

```python
employes={"employee_name":['John', 'Kate', 'Leo'], "department":['IT', 'HR', 'Sales'],"salary":[50000, 55000, 60000]}
employe_frame=pd.DataFrame(employes)
print(employe_frame)
```

```python
data=[['John', 'Kate', 'Leo'],
    ['IT', 'HR', 'Sales'],
    [50000, 55000, 60000]]
series_data=pd.Series(data)
print(series_data)
```

```python
data=[['John', 'Kate', 'Leo'],
    ['IT', 'HR', 'Sales'],
    [50000, 55000, 60000]]

frame_data=pd.DataFrame(data)
print(frame_data)
```
