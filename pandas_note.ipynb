Pandas:
Pandas is python library used for data manipulation and data analysis. 
Pandas provides a convenient way to analyze and clean data.
Pandas is mostly used for 2D arrays. i.e rows and columns. 2D array is in tabular format.
Pandas has two popular data types. They are :-

Series :- 
series have only one column. 
It is a one dimensional array.
It is more faster then data Frame.
we do not use series that much because in any dataset their are more than 1 columns.


Data frame :- 
data frame can have one or multiple columns. 
It is multi dimensional array.
It is also called as the collection of series.
Data frame is slower than series.


```python
# creating pandas series:
import pandas as pd
list1=[1,2,3,"string"]
list2=[5,6,7,"abcd"]
list3=list1+list2
print(list3)
series=pd.Series(list3) # list3 is represented in a column
print(series)
```

```python
# if we have to give different index then default index. i.e 0,1,2,3,4,5.......
series1=pd.Series(list3, index=['a','b','c','d','e','f','g','h'])
print(series1)
```

Remember the length of list and length of index must be equal otherwise error will occur.


```python
# create a tuple
data= (10,20,30,40,50,60,70)
# create a series from the tuple
my_series = pd.Series(data)
print(my_series)
```

```python
# create a dictionary
grades = {"semester1":['john','ram','hari'],"semester2":3.35,"semester3":2.8}
#grades = [3.14, 2.98, 3.04]
# create a series from the dictionary
my_series = pd.Series (grades)
print(my_series)
```

In dictionary, series treat keyword as index and entire value of that keyword as its single column data.


```python
# in series their an attribute called .index which gives all index of that series.
print(my_series.index)  # gives index
print()
print(my_series.values) # gives values
```

```python
s=pd.Series([1,2,3,4,5,6,7])
print(s)
```

```python
# creating dataframe

data = {"Name": ['John', 'Alice', 'Bob'], 'Age': [25, 28, 32], 'City': ['New York', 'London', 'Paris']}

df = pd.DataFrame(data)
print(df)
```

In dataframe key becomes column head and values of that key becomes its column data as above

data = {'Name': ['John', 'Alice', 'Bob','Rajan'], 
        'Age': [25, 28, 32],

```python
import numpy as np

data = {'Name': ['John', 'Alice', 'Bob','Rajan'], 
        'Age': [25, 28, 32,np.nan], 
        'City': ['New York', 'London', 'Paris',np.nan]}
tabular_data=pd.DataFrame(data)
print(tabular_data)
```

we can use np.nan to fill with Nan(null value). it is under numpy so we have to import numpy

```python
data = {'Name': ['John', 'Alice', 'Bob'], 
        'Age': [25, 28, 32], 
        'City': ['New York', 'London', 'Paris']}

df=pd.DataFrame(data,index=pd.RangeIndex(10,13,name='s.n'))
print(df)
```

df=pd.DataFrame(data,index=pd.RangeIndex(10,13,name='s.n'))

above code is used if we need to set the index starting, ending and index name

Note: length of index must be of equal range 

```python
# creating data with list of list 
data=[['John', 'Kate', 'Leo'],
    ['IT', 'HR', 'Sales'],
    [50000, 55000, 60000]]

frame_data=pd.DataFrame(data)
print(frame_data)
```

it have default index and also column name as default index


```python
data=[['John', 25, 'New York'],
    ['Kate', 30, 'London'],
    ['Bob', 28, 'Paris']]

frame_data=pd.DataFrame(data)
frame_data=pd.DataFrame(data,columns=['Name','Age','City'])
print(frame_data)
```

frame_data=pd.DataFrame(data,columns=['Name','Age','City'])
